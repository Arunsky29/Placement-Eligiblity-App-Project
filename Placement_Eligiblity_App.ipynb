```python
from faker import Faker
import random
import sqlite3
import pandas as pd
```

```python
fake = Faker()
Faker.seed(0)
random.seed(0)
```

#CREATING TABLES

```python
NUM_STUDENTS = 500
genders = ['Male', 'Female', 'Other']
batches = ['ECE2021', 'EEE2022', 'EEE2021', 'ECE2022']
placement_status_options = ['Ready', 'Not Ready', 'Placed']
companies = ['Google', 'Amazon', 'TCS', 'Infosys', 'Capgemini', 'Flipkart', 'Wipro', None]
```

```python
students_data = []

for student_id in range(1, NUM_STUDENTS + 1):
    name = fake.name()
    age = random.randint(20, 26)
    gender = random.choice(genders)
    email = fake.email()
    phone = fake.phone_number()
    enrollment_year = random.randint(2019, 2023)
    course_batch = random.choice(batches)
    city = fake.city()
    graduation_year = enrollment_year + 3

    students_data.append((
        student_id, name, age, gender, email, phone,
        enrollment_year, course_batch, city, graduation_year
    ))
```

```python
df_students = pd.DataFrame(students_data, columns=[
    "student_id", "name", "age", "gender", "email", "phone",
    "enrollment_year", "course_batch", "city", "graduation_year"
])
df_students.sample(5)
```

```python
languages = ['Python', 'SQL', 'Java', 'C++']

programming_data = []

for prog_id in range(1, NUM_STUDENTS + 1):
    student_id = prog_id  # 1-to-1 mapping
    language = random.choice(languages)
    problems_solved = random.randint(10, 200)
    assessments_completed = random.randint(1, 10)
    mini_projects = random.randint(0, 5)
    certifications_earned = random.randint(0, 3)
    latest_project_score = round(random.uniform(40, 100), 2)

    programming_data.append((
        prog_id, student_id, language, problems_solved,
        assessments_completed, mini_projects,
        certifications_earned, latest_project_score
    ))
```

```python
df_programming = pd.DataFrame(programming_data, columns=[
    "programming_id", "student_id", "language", "problems_solved",
    "assessments_completed", "mini_projects", "certifications_earned",
    "latest_project_score"
])
df_programming.sample(5)
```

```python
soft_skills_data = []

for soft_id in range(1, NUM_STUDENTS + 1):
    student_id = soft_id  # 1-to-1 mapping
    communication = random.randint(40, 100)
    teamwork = random.randint(40, 100)
    presentation = random.randint(40, 100)
    leadership = random.randint(40, 100)
    critical_thinking = random.randint(40, 100)
    interpersonal_skills = random.randint(40, 100)

    soft_skills_data.append((
        soft_id, student_id, communication, teamwork,
        presentation, leadership, critical_thinking, interpersonal_skills
    ))
```

```python
df_soft_skills = pd.DataFrame(soft_skills_data, columns=[
    "soft_skill_id", "student_id", "communication", "teamwork",
    "presentation", "leadership", "critical_thinking", "interpersonal_skills"
])
df_soft_skills.sample(5)
```

```python
placements_data = []

for place_id in range(1, NUM_STUDENTS + 1):
    student_id = place_id
    mock_score = random.randint(30, 100)
    internships = random.randint(0, 3)
    status = random.choices(
        placement_status_options, weights=[0.4, 0.3, 0.3], k=1
    )[0]

    if status == "Placed":
        company = random.choice(companies[:-1])
        package = round(random.uniform(3.5, 20), 2)  # In LPA or similar
        rounds = random.randint(2, 6)
        date = fake.date_between(start_date='-1y', end_date='today')
    else:
        company = None
        package = None
        rounds = None
        date = None

    placements_data.append((
        place_id, student_id, mock_score, internships,
        status, company, package, rounds, date
    ))
```

```python
df_placements = pd.DataFrame(placements_data, columns=[
    "placement_id", "student_id", "mock_interview_score", "internships_completed",
    "placement_status", "company_name", "placement_package",
    "interview_rounds_cleared", "placement_date"
])
df_placements.sample(5)
```

```python
import mysql.connector

class PlacementDBManagerTiDB:
    def __init__(self, host, user, password, database):
        self.conn = mysql.connector.connect(
            host=host,
            user=user,
            password=password,
            database=database
        )
        self.cursor = self.conn.cursor()

    def create_tables(self):
        self.cursor.execute("""
        CREATE TABLE IF NOT EXISTS Students (
            student_id INT PRIMARY KEY,
            name VARCHAR(100),
            age INT,
            gender VARCHAR(10),
            email VARCHAR(100),
            phone VARCHAR(100),
            enrollment_year INT,
            course_batch VARCHAR(20),
            city VARCHAR(100),
            graduation_year INT
        )""")

        self.cursor.execute("""
        CREATE TABLE IF NOT EXISTS Programming (
            programming_id INT PRIMARY KEY,
            student_id INT,
            language VARCHAR(20),
            problems_solved INT,
            assessments_completed INT,
            mini_projects INT,
            certifications_earned INT,
            latest_project_score FLOAT,
            FOREIGN KEY (student_id) REFERENCES Students(student_id)
        )""")

        self.cursor.execute("""
        CREATE TABLE IF NOT EXISTS SoftSkills (
            soft_skill_id INT PRIMARY KEY,
            student_id INT,
            communication INT,
            teamwork INT,
            presentation INT,
            leadership INT,
            critical_thinking INT,
            interpersonal_skills INT,
            FOREIGN KEY (student_id) REFERENCES Students(student_id)
        )""")

        self.cursor.execute("""
        CREATE TABLE IF NOT EXISTS Placements (
            placement_id INT PRIMARY KEY,
            student_id INT,
            mock_interview_score INT,
            internships_completed INT,
            placement_status VARCHAR(20),
            company_name VARCHAR(100),
            placement_package FLOAT,
            interview_rounds_cleared INT,
            placement_date DATE,
            FOREIGN KEY (student_id) REFERENCES Students(student_id)
        )""")

        self.conn.commit()

    def insert_data(self, table_name, data):
        placeholders = ",".join(["%s"] * len(data[0]))
        query = f"INSERT INTO {table_name} VALUES ({placeholders})"
        self.cursor.executemany(query, data)
        self.conn.commit()

    def close(self):
        self.conn.close()
```
